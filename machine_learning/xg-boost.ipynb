## Importing the Libraries

````python
import pandas as pd
from numpy import loadtxt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
````

## Loading the Dataset

````python
path = '../datasets/pima-indians-diabetes.csv'
dataset = loadtxt(path, delimiter = ',')
````

````python
print(dataset)
````

## Data Preparation

````python
X = dataset[:, 0:8]
y = dataset[:, 8]
````

## Train Test Split

````python
seed = 7
test_size = 0.33

X_train, X_test, y_train, y_test = train_test_split(
 X, y, 
 test_size = test_size,
 random_state = seed
)
````

## Model Creation

````python
model = XGBClassifier()
````

## Model Training

````python
model.fit(X_train, y_train)
````

## Model Prediction

````python
predictions = model.predict(X_test)
````

````python
print(predictions)
````

## Model Accuracy

````python
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
````

## Iris Dataset
---

### Load the dataset

````python
path = "../datasets/iris.csv"
df2 = pd.read_csv(path, header=None)
````

````python
df2.head()
````

### Data Preperation

````python
data = df2.values
print(data[0:5])
````

````python
X = data[:,0:4]
Y = data[:,4]
````

````python
print(X[0:5])
print()
print(Y[0:5])
````

### Encoding the target values

````python
le = LabelEncoder()
le = le.fit(Y)
le_y = le.transform(Y)
````

````python
print(le_y)
````

### Train Test Split

````python
seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    le_y, 
    test_size=test_size, 
    random_state = seed
    )
````

### Model Creation

````python
model = XGBClassifier()
````

### Model Training

````python
model.fit(X_train, y_train)
````

### Model Prediction

````python
predictions = model.predict(X_test)
print(X_test[0:5])
print()
print(predictions[0:5])
````

### Model Evaluation

````python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
````

# Feature Importance using XGBoost

* XGBoost helps in Feature selection as well.
* Based on this score.. we can select the best features for the model

````python
# Load the dataset
dataset = loadtxt('../datasets/pima-indians-diabetes.csv', delimiter = ',')

# Data Preparation
X = dataset[:, 0:8]
y = dataset[:, 8]

# Model creation
model = XGBClassifier()

# Model Training
model.fit(X, y)

# Import Features extraction
print(model.feature_importances_)
````

````python
r = range(len(model.feature_importances_))
f_imp = model.feature_importances_

plt.bar(r, f_imp)
plt.show()
````
